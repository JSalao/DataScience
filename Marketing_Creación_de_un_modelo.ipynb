# Entrenamiento y ejecución de su primer modelo

Hemos aprendido que los modelos son códigos informáticos que procesan información para hacer una predicción o tomar una decisión. 
Aquí vamos a entrenar un modelo para que adivine una talla de bota cómoda para un perro, basándose en la talla del arnés que le corresponde.

image.png

```python
import pandas as pd
import statsmodels
```

```python
dataset = pd.read_csv("arnes.csv")
```

```python
# Print the data
dataset
```

Como puede ver, tenemos las tallas de botas y arneses para 50 perros de nieve.

Queremos utilizar el tamaño del arnés para estimar el tamaño de la bota. Esto significa que harness_size es nuestra entrada. 

Queremos un modelo que procese la entrada y haga sus propias estimaciones del tamaño de la bota (salida).

**Seleccionar un modelo**


Lo primero que debemos hacer es seleccionar un modelo. Estamos empezando, así que comenzaremos con un modelo muy simple llamado OLS. Esto es sólo una línea recta (a veces llamada línea de tendencia).

Vamos a utilizar una biblioteca existente para crear nuestro modelo, pero no lo entrenaremos todavía

```python
# Cargar una biblioteca para que haga el trabajo duro por nosotros
import statsmodels.formula.api as smf

# Primero, definimos nuestra fórmula usando una sintaxis especial
# Esto dice que boot_size se explica por harness_size
formula = "boot_size ~ harness_size"

# Creamos el modelo, pero no lo entrenamos todavía
model = smf.ols(formula = formula, data = dataset)
```

**Entrenamiento de nuestro modelo**


Los modelos OLS tienen dos parámetros (una pendiente y un desplazamiento), pero aún no se han establecido en nuestro modelo. 

Tenemos que entrenar (ajustar) nuestro modelo para encontrar estos valores, de modo que el modelo pueda estimar de forma fiable el tamaño de las botas de los perros en función del tamaño de su arnés.

El siguiente código ajusta nuestro modelo a los datos que ha visto ahora

```python
# Train (fit) the model so that it creates a line that 
# fits our data. This method does the hard work for
# us.
fitted_model = model.fit()
print("Modelo Entrenado")
```

**Uso del modelo**
Ahora que hemos terminado el entrenamiento, podemos utilizar nuestro modelo para predecir el tamaño de las botas de un perro a partir de su tamaño de arnés.

```python
# harness_size indica el tamaño del arnés que nos interesa
harness_size = { 'harness_size' : [50] }

# Utiliza el modelo para predecir la talla de las botas del perro
approximate_boot_size = fitted_model.predict(harness_size)

# Print the result
print("Tamaño aproximado de la bota:")
print(approximate_boot_size[0])
```
